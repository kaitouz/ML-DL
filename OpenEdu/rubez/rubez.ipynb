```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('pulsar_stars_new.csv')
df.head()
```

```python
# Filter, select the rows that satisfy the following conditions:
# + TG = 0 & 90.3515625 <= MIP <= 91.234375
# or
# + TG = 1 & 20.7890625 <= MIP <= 25.890625

df_droped = df.drop(df[ ((df['TG'] == 0) & ((df['MIP'] < 90.3515625) | (df['MIP'] > 91.234375)))  
                      | ((df['TG'] == 1) & ((df['MIP'] < 20.7890625) | (df['MIP'] > 25.890625)))].index)

# Ques#1: Count the number of rows left after filtering 
# Ques#2: Calculate mean of column MIP 
# Ques#3: Caculate max of column SIP
df_droped.describe()
```

```python
# Выполните линейную нормировку всех значений предикторов полученной выборки. (Implement linear normalization for Predictors)
```

```python
# Predictors
X_train = df_droped.drop('TG', axis=1)
X_train
```

```python
# Output
y_train = df_droped['TG']
y_train.head(10)
```

```python
# Linear normalization: x_norm = (x - x_min)/(x_max - x_min) (Or use sklearn.preprocessing.MinMaxScaler)
X_norm = (X_train - X_train.min()) / (X_train.max() - X_train.min())

# Ques#4: Calculate mean of column MIP after normaliztion
X_norm.describe()
```

```python
# 
new_obj = [[0.598, 0.748, 0.809, 0.913, 0.667, 0.608, 0.473, 0.731]]
```

```python
# Train a logistic regression model (using normalized data).

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=2019, solver='lbfgs')
model.fit(np.array(X_norm), y_train)
```

```python
# Ques#5: Caculate probability of classifying the star as a pulsar: (TG = 1) (e.g. result = 0.82449167)
model.predict_proba(new_obj)
```

```python
# Ques#6: Caculate the distance from the new star to its nearest neighbor using the Euclidean metric.

from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=1, p=2) # p = 2 because using the Euclidean metric
neigh.fit(np.array(X_norm), y_train)  
neigh.kneighbors(new_obj)  # e.g. result = 1.179112 (= neigh.kneighbors(new_obj)[0][0][0])
```
