# INPUT

```python
# Input
file_name = 'data.csv'
str_remove = 'Магаданская обл., г. Санкт-Петербург, Тульская область'

# question 1
query = [1, 41, 72] 
```

# Processing

```python
import pandas as pd
from pandas import *
import numpy as np
```

```python
df = pd.read_csv(file_name)
df
```

```python
list_remove = str_remove.split(', ')
list_remove
```

```python
df_removed = df.drop(df[df['region_name'].isin(list_remove)].index)
df_removed
```

```python
df_sorted = df_removed.sort_values('salary')
df_sorted
```

```python
df_sorted = df_sorted.reset_index().drop(['index'], axis = 1)
df_sorted.index += 1
df_sorted
```

# Result

```python
print(f'X({query[0]}) = {df_sorted.loc[query[0]].salary}')
print(f'X({query[1]}) = {df_sorted.loc[query[1]].salary}')
print(f'X({query[2]}) = {df_sorted.loc[query[2]].salary}\n')

print(f'Mean = {df_sorted.salary.mean()}')
print(f'Median = {df_sorted.salary.median()}')
```
