```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img1 = cv2.imread('1.jpg')
RGB_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img1)
```

```python
gray_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img1, cmap='gray')
```

```python
gray_hist = plt.hist(gray_img1.ravel(), bins = 256)
plt.xlabel('Интенсивность $n$')
plt.ylabel('Количество $h(n)$')
plt.show()
```

```python
li = np.array([np.sum([gray_img1.ravel() == l]) for l in range(0, 256)])
```

```python
print(li.max(), li.argmax())
```

```python
img2 = cv2.imread('2.jpg')
RGB_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img2)
```

```python
plt.hist(RGB_img2[:, :, 0].ravel(), bins = 256, color = 'Red', alpha = 0.5, density=True)
plt.hist(RGB_img2[:, :, 1].ravel(), bins = 256, color = 'Green', alpha = 0.5, density=True)
plt.hist(RGB_img2[:, :, 2].ravel(), bins = 256, color = 'Blue', alpha = 0.5, density=True)
plt.legend(['Красный канал (R)', 'Зеленый канал (G)', 'Cиний канал (B)'])
plt.xlabel('Интенсивность $n$')
plt.ylabel('Количество')
plt.show()
```

```python
li2 = np.array([np.sum(RGB_img2[:, :, 2].ravel() == i) for i in range(256)])
li2.max()
```

```python
x, y, _ = plt.hist(RGB_img2[:, :, 2].ravel(), bins = 256, color = 'Red', alpha = 0.5, density=True)
x.max()
```
