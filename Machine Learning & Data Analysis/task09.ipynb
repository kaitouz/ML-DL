```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv",index_col="Object")
df
```

```python
df_task = df.drop('Cluster', axis=1)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, init=np.array([[5.5, 11.5], [10.57, 7.57], [7.5, 15.0]]), max_iter=100, n_init=1)
```

```python
model = kmeans.fit(df_task)
```

```python
df_ans = df_task.copy()
df_ans['Clusters'] = model.labels_.tolist()
df_ans
```

```python
df_task
```

```python
alldistances = kmeans.fit_transform(df_task)
alldistances
```

```python
search = 0
Clusters_array = []
for i in range(len(df_ans['Clusters'])):
  if (df_ans['Clusters'][i+1] == search):
    Clusters_array.append(alldistances[i][search])
sum(Clusters_array)/len(Clusters_array)
```
